**Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**load the processed datasets created in the notebook Part I where** 
1. **a** is a train dataset and 
2. **b** is a test dataset

```python
a_categorical=pd.read_csv('train_processed_continous.csv')
b_categorical=pd.read_csv('test_processed_continous.csv')
a_categorical.head()
```

```python
a_categorical.shape,b_categorical.shape
```

**converting the Age column datatype into (int) and assiging the Age range into 5 parts**

```python
combine=[a_categorical,b_categorical]
for dataset in combine:
    dataset['Age']=dataset['Age'].astype(int)
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4
a_categorical.head()
```

**Assiging the Fare range into 6 parts**

```python
combine = [a_categorical, b_categorical]

for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[(dataset['Fare'] > 31) & (dataset['Fare'] <= 99), 'Fare']   = 3
    dataset.loc[(dataset['Fare'] > 99) & (dataset['Fare'] <= 250), 'Fare']   = 4
    dataset.loc[ dataset['Fare'] > 250, 'Fare'] = 5
    dataset['Fare'] = dataset['Fare'].astype(int)
```

```python
a_categorical
```

**combining the 'SibSp','Parch' and forming the 'relative' column ,creating another feature 'not_alone' from 'relative'
if the number relatives is zero 'not_alone' column takes 0 else it takes 1**

```python
combine = [a_categorical, b_categorical]
for dataset in combine:
    dataset['relatives'] = dataset['SibSp'] + dataset['Parch']
    dataset.loc[dataset['relatives'] > 0, 'not_alone'] = 0
    dataset.loc[dataset['relatives'] == 0, 'not_alone'] = 1
    dataset['not_alone'] = dataset['not_alone'].astype(int)
```

**droping the 'Survived','Parch','SibSp' column in a2 and 'PassengerId','Parch','SibSp' column in b1**

```python
a2=a_categorical.drop(['Survived','Parch','SibSp'],axis=1)
b1=b_categorical.drop(['PassengerId','Parch','SibSp'],axis=1)
a2.shape,b1.shape
```

**we are trying to find the maximum survival rate so for that we are creating new columns**

```python
combine=[a2,b1]
for dataset in combine:
    dataset['Fare_Per_Person'] = dataset['Fare']/(dataset['relatives']+1)
    dataset['Fare_Per_Person'] = dataset['Fare_Per_Person'].astype(int)
```

```python
X_train = a2_encoded
Y_train = a_categorical["Survived"]
X_test  = b1_encoded
```

```python
# Algorithms
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_train,Y_train,test_size=0.3,random_state=1,shuffle=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# stochastic gradient descent (SGD) learning
sgd = linear_model.SGDClassifier(max_iter=5, tol=None)    #number of times trained in dataset(max_iter=5)
sgd.fit(x_train, y_train)                                   
Y_pred = sgd.predict(x_test)

acc_sgd = round(accuracy_score(y_test, Y_pred) * 100, 2)

print(round(acc_sgd,2,), "%")
```

```python
sgd.score(x_test,y_test)   #dataset(X_train,Y_train) where parch and sibsp are taken continous
```

```python
# Random Forest
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)                        

Y_prediction = random_forest.predict(x_test)


acc_random_forest = round(accuracy_score(y_test,Y_prediction) * 100, 2)
print(round(acc_random_forest,2,), "%")
```

```python
random_forest.score(x_test,y_test)         
```

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)            

Y_pred = logreg.predict(x_test)

acc_log = round(accuracy_score(y_test, Y_pred) * 100, 2)
print(round(acc_log,2,), "%")
```
